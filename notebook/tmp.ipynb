```python
import mlflow

# 1. Initialize MLflow client
client = mlflow.tracking.MlflowClient()

# 2. Run ID from which to remove the 'registered' status
run_id = "6ffbb8dafe5e4395b1c93d2547ad160a"

# 3. Delete the tag containing the model registration info.
#    This tag is responsible for displaying the 'registered' icon.
try:
    client.delete_tag(run_id, "mlflow.registered_model_name")
    print(f"Successfully deleted 'mlflow.registered_model_name' tag from Run ID {run_id}.")
    print("You can see the status change by refreshing the MLflow UI.")

    # Sometimes this tag may also need to be deleted.
    client.delete_tag(run_id, "mlflow.log-model.history")

except Exception as e:
    print(f"Error while deleting tag: {e}")
```
